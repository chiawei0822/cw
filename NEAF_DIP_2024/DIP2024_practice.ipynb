## 1. Histogram Equalization
Histogram is the method to enhance image contrast.
Please show the image grayscale histogram and do the equlization, and show the image result and histogram equalization result.

```python
img_ori = cv2.imread("input\lena.bmp", cv2.IMREAD_GRAYSCALE) # Read image as gray.

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.imshow(img_ori,cmap='gray')
plt.title("origin")
plt.axis("off")

hist1 = cv2.calcHist([img_ori], [0], None, [256], [0, 256])
plt.subplot(2,2,2)
plt.plot(hist1)

equalize_img = cv2.equalizeHist(img_ori)
plt.subplot(2,2,3)
plt.imshow(equalize_img,cmap='gray')
plt.title("equalize")
plt.axis("off")

hist2 = cv2.calcHist([equalize_img], [0], None, [256], [0, 256])
plt.subplot(2,2,4)
plt.plot(hist2)

# save result
cv2.imwrite('output/img_ori.bmp', img_ori)
cv2.imwrite('output/img_equa.bmp', equalize_img)
```

## 2.Mean blur, Gaussian blur, Median blur
Blur is a method to deal with the noise on the image. Please input the image and use different kernel size to show blur result.

```python
img_noise = cv2.imread("input\lena_noise.png", cv2.IMREAD_GRAYSCALE)
plt.subplot(1,4,1)
plt.imshow(img_noise,cmap='gray')
plt.title("origin")
plt.axis("off")

# Blur the image by three methods.
image_meanBlur = cv2.blur(img_noise, (5,5))
image_GaussianBlur = cv2.GaussianBlur(img_noise, (5,5),0)
image_medianBlur = cv2.medianBlur(img_noise, 5)

#save result
cv2.imwrite('output/img_mean_blur.png', image_meanBlur)
cv2.imwrite('output/img_gaussian_blur.png', image_GaussianBlur)
cv2.imwrite('output/img_median_blur.png', image_medianBlur)

plt.subplot(1,4,2)
plt.imshow(image_meanBlur,cmap='gray')
plt.title("meanBlur")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(image_GaussianBlur,cmap='gray')
plt.title("GaussianBlur")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(image_medianBlur,cmap='gray')
plt.title("medianBlur")
plt.axis("off")
```

## 5. Morphological: erosion, dilation, open, close

Those methods help to denoise, enhance the features.   
Please use these methods to fill up the lines of the image created in the previous section.

```python
hsv_yellow = cv2.imread('output/cropHSV_yellow.jpg')
plt.figure(figsize=(20,6.5))
plt.subplot(2,5,1)
plt.imshow(hsv_yellow,cmap='gray')
plt.title("origin")
plt.axis("off")

hist1 = cv2.calcHist([hsv_yellow], [0], None, [256], [0, 256])
plt.subplot(2,5,2)
plt.plot(hist1)

_, image_binary = cv2.threshold(hsv_yellow, 130, 255, cv2.THRESH_BINARY) ## (threshold, image)
gray = cv2.cvtColor(image_binary, cv2.COLOR_BGR2GRAY)
print("_ = ", _)
plt.subplot(2,5,3)
plt.imshow(gray,cmap="gray")
plt.title("threshold")
plt.axis("off")

hist2 = cv2.calcHist([gray], [0], None, [256], [0, 256])
plt.subplot(2,5,4)
plt.plot(hist2)


# Try to use erosion, dilation, open, close to fill up the lines.

kernel1 = np.ones((5, 5), np.uint8)
dilated_img = cv2.dilate(gray, kernel1, iterations=1)
plt.subplot(2,5,5)
plt.imshow(dilated_img,cmap="gray")
plt.title("dilated_img")
plt.axis("off")

kernel2 = np.ones((5, 5), np.uint8)
eroded_img = cv2.erode(dilated_img, kernel2, iterations=1)
plt.subplot(2,5,6)
plt.imshow(eroded_img,cmap="gray")
plt.title("eroded_img")
plt.axis("off")

kernel4 = np.ones((2, 2), np.uint8)
dilated_img2 = cv2.dilate(eroded_img, kernel4, iterations=1)
plt.subplot(2,5,7)
plt.imshow(dilated_img2,cmap="gray")
plt.title("dilated_img2")
plt.axis("off")

kernel3 = np.ones((4, 4), np.uint8)
eroded_img2 = cv2.erode(dilated_img2, kernel3, iterations=1)
plt.subplot(2,5,8)
plt.imshow(eroded_img2,cmap="gray")
plt.title("eroded_img2")
plt.axis("off")

kernel5 = np.ones((3, 3), np.uint8)
dilated_img3 = cv2.dilate(eroded_img2, kernel5, iterations=1)
plt.subplot(2,5,9)
plt.imshow(dilated_img3,cmap="gray")
plt.title("dilated_img3")
plt.axis("off")

# save result
cv2.imwrite('output/morpho_yellow.jpg', dilated_img3)
```
