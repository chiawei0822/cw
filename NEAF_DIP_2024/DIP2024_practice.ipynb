## 1. Histogram Equalization
Histogram is the method to enhance image contrast.
Please show the image grayscale histogram and do the equlization, and show the image result and histogram equalization result.

```python
img_ori = cv2.imread("input\lena.bmp", cv2.IMREAD_GRAYSCALE) # Read image as gray.

plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.imshow(img_ori,cmap='gray')
plt.title("origin")
plt.axis("off")

hist1 = cv2.calcHist([img_ori], [0], None, [256], [0, 256])
plt.subplot(2,2,2)
plt.plot(hist1)

equalize_img = cv2.equalizeHist(img_ori)
plt.subplot(2,2,3)
plt.imshow(equalize_img,cmap='gray')
plt.title("equalize")
plt.axis("off")

hist2 = cv2.calcHist([equalize_img], [0], None, [256], [0, 256])
plt.subplot(2,2,4)
plt.plot(hist2)

# save result
cv2.imwrite('output/img_ori.bmp', img_ori)
cv2.imwrite('output/img_equa.bmp', equalize_img)
```

## 2.Mean blur, Gaussian blur, Median blur
Blur is a method to deal with the noise on the image. Please input the image and use different kernel size to show blur result.

```python
img_noise = cv2.imread("input\lena_noise.png", cv2.IMREAD_GRAYSCALE)
plt.subplot(1,4,1)
plt.imshow(img_noise,cmap='gray')
plt.title("origin")
plt.axis("off")

# Blur the image by three methods.
image_meanBlur = cv2.blur(img_noise, (5,5))
image_GaussianBlur = cv2.GaussianBlur(img_noise, (5,5),0)
image_medianBlur = cv2.medianBlur(img_noise, 5)

#save result
cv2.imwrite('output/img_mean_blur.png', image_meanBlur)
cv2.imwrite('output/img_gaussian_blur.png', image_GaussianBlur)
cv2.imwrite('output/img_median_blur.png', image_medianBlur)

plt.subplot(1,4,2)
plt.imshow(image_meanBlur,cmap='gray')
plt.title("meanBlur")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(image_GaussianBlur,cmap='gray')
plt.title("GaussianBlur")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(image_medianBlur,cmap='gray')
plt.title("medianBlur")
plt.axis("off")
```

## 4. HSV filter
This method could extract the the interested part by color.  
Please crop the yellow or blue boxes and its surrounding in the image.

```python
import numpy as np

# Yellow box.
yellow = cv2.imread('input/yellow.jpg')

# Step 1. Transfer color to HSV
image_hsv = cv2.cvtColor(yellow, cv2.COLOR_BGR2HSV)

# Step 2. Setting the color range and filter the image.
# lowerb = np.array([26,43,46])
# upperb = np.array([99,255,255])
lowerb = np.array([30,100,100])
upperb = np.array([34,255,255])
mask = cv2.inRange(image_hsv, lowerb, upperb)
# print(mask)
x_arr = []; y_arr=[]
for i in range(len(mask)):
    for j in range(len(mask[0])):
        if mask[i][j] != 0:
            # print(i,j,mask[i][j])
            x_arr.append(i)
            y_arr.append(j)

x_max = max(x_arr)
x_min = min(x_arr)
y_max = max(y_arr)
y_min = min(y_arr)
print(x_max, x_min, y_max, y_min)

image_masked = cv2.bitwise_and(yellow, yellow, mask=mask)
image_yellow = cv2.cvtColor(yellow, cv2.COLOR_BGR2RGB)

# crop_edge = 20  # surrounding 20 pixels 
# x = 3170
# y = 2520
# w = 30
# h = 30
# cropmask = image_masked[y-crop_edge: y+h+crop_edge, x-crop_edge: x+w+crop_edge]   # take (x, y, w, h) as example.
# plt.imshow(cropmask)
# plt.title("crop")
# plt.axis("off")

image_list = [yellow, image_hsv, mask, image_masked, image_yellow]
title_list = ['yellow', 'image_hsv', 'mask', 'image_masked', 'image_yellow']
plt.figure(figsize=(40,8))
for i in range(len(title_list)):
    plt.subplot(2,5,i+1)
    plt.imshow(image_list[i],cmap="gray")
    plt.title(title_list[i],size=50)
    plt.axis("off")

# Step 3. Binary the image, find the left-top and right-bottom coordinate (x, y, w, h) or (x1, x2, y1, y2).
hist1 = cv2.calcHist([image_masked], [0], None, [256], [0, 256])
plt.subplot(2,5,6)
plt.plot(hist1)
_, image_binary = cv2.threshold(image_masked, 50, 255, cv2.THRESH_BINARY) ## (threshold, image)
print("_ = ", _)
plt.subplot(2,5,7)
plt.imshow(image_binary,cmap="gray")
plt.title("threshold")
plt.axis("off")

crop_edge = 20  # surrounding 20 pixels 
x = 3170
y = 2520
w = 30
h = 30
cropthres = image_binary[y-crop_edge: y+h+crop_edge, x-crop_edge: x+w+crop_edge]   # take (x, y, w, h) as example.
plt.subplot(2,5,8)
plt.imshow(cropthres,cmap="gray")
plt.title("threshold_crop")
plt.axis("off")

gray = cv2.cvtColor(image_binary, cv2.COLOR_BGR2GRAY)
contours_yellow, _ = cv2.findContours(gray, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
# print("max=", np.max(contours_yellow))
# print(_)
contour_image = yellow.copy()
cv2.drawContours(contour_image, contours_yellow, -1, (0, 255, 255), 2)
crop_edge = 20  # surrounding 20 pixels 
x = 3170
y = 2520
w = 30
h = 30
cropcounter = contour_image[y-crop_edge: y+h+crop_edge, x-crop_edge: x+w+crop_edge]   # take (x, y, w, h) as example.
plt.subplot(2,5,9)
plt.imshow(cropcounter)
plt.title("counter")
plt.axis("off")


# Step 4. Crop the image by the coordinate.

crop_edge = 20  # surrounding 20 pixels 
x = y_min
y = x_min
h = x_max-x_min
w = y_max-y_min
cropHSV_yellow = image_yellow[y-crop_edge: y+h+crop_edge, x-crop_edge: x+w+crop_edge]   # take (x, y, w, h) as example.

plt.subplot(2,5,10)
plt.imshow(cropHSV_yellow)
plt.title("crop")
plt.axis("off")

# save result
cv2.imwrite('output/cropHSV_yellow.jpg', cropHSV_yellow)
```

```python
# Blue box.
# blue = cv2.imread('input/blue.jpg')

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
blue_image = cv2.imread('input/blue.jpg')

# Step 1. Convert the color to HSV
image_hsv = cv2.cvtColor(blue_image, cv2.COLOR_BGR2HSV)

# Step 2. Set the color range for blue and filter the image
# These ranges are typical for detecting blue colors, you might need to adjust based on the specific image.
lowerb = np.array([100, 150, 150])
upperb = np.array([140, 255, 255])
mask = cv2.inRange(image_hsv, lowerb, upperb)

# Check if the mask correctly isolates the blue area
plt.figure(figsize=(10,5))
plt.subplot(1, 3, 1)
plt.imshow(mask, cmap='gray')
plt.title("Mask")
plt.axis("off")

# Step 3. Find contours directly on the mask
contours_blue, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
contour_image = blue_image.copy()
cv2.drawContours(contour_image, contours_blue, -1, (255, 0, 0), 2)

# Display the contour result
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(contour_image, cv2.COLOR_BGR2RGB))
plt.title("Contours")
plt.axis("off")

# Step 4. Find bounding box of the largest contour and crop the image
if contours_blue:
    x, y, w, h = cv2.boundingRect(contours_blue[0])
    crop_edge = 65  # Adding some padding
    cropHSV_blue = blue_image[max(y-crop_edge, 0):y+h+crop_edge, max(x-crop_edge, 0):x+w+crop_edge]

    plt.subplot(1, 3, 3)
    plt.imshow(cv2.cvtColor(cropHSV_blue, cv2.COLOR_BGR2RGB))
    plt.title("Cropped Image")
    plt.axis("off")

    # Save result
    cv2.imwrite('/mnt/data/cropHSV_blue.jpg', cropHSV_blue)
else:
    print("No blue region detected.")

plt.show()


# save result
cv2.imwrite('output/cropHSV_blue.jpg', cropHSV_blue)
```

## 5. Morphological: erosion, dilation, open, close

Those methods help to denoise, enhance the features.   
Please use these methods to fill up the lines of the image created in the previous section.

```python
hsv_yellow = cv2.imread('output/cropHSV_yellow.jpg')
plt.figure(figsize=(20,6.5))
plt.subplot(2,4,1)
plt.imshow(hsv_yellow,cmap='gray')
plt.title("origin")
plt.axis("off")

hist1 = cv2.calcHist([hsv_yellow], [0], None, [256], [0, 256])
plt.subplot(2,4,2)
plt.plot(hist1)

gray = cv2.cvtColor(hsv_yellow, cv2.COLOR_BGR2GRAY)
_, image_binary = cv2.threshold(gray, 110, 255, cv2.THRESH_BINARY) ## (threshold, image)

print("_ = ", _)
plt.subplot(2,4,3)
plt.imshow(image_binary,cmap="gray")
plt.title("threshold")
plt.axis("off")

hist2 = cv2.calcHist([image_binary], [0], None, [256], [0, 256])
plt.subplot(2,4,4)
plt.plot(hist2)


# Try to use erosion, dilation, open, close to fill up the lines.

kernel = np.ones((5, 5), np.uint8)
dilated_img = cv2.dilate(image_binary, kernel, iterations=1)
plt.subplot(2,4,5)
plt.imshow(dilated_img,cmap="gray")
plt.title("dilated_img")
plt.axis("off")

eroded_img = cv2.erode(dilated_img, kernel, iterations=1)
plt.subplot(2,4,7)
plt.imshow(eroded_img,cmap="gray")
plt.title("eroded_img")
plt.axis("off")

# save result
cv2.imwrite('output/morpho_yellow.jpg', eroded_img)
```

```python
hsv_blue = cv2.imread('output/cropHSV_blue.jpg', cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(20,6.5))
plt.subplot(2,4,1)
plt.imshow(hsv_blue,cmap='gray')
plt.title("origin")
plt.axis("off")

hist1 = cv2.calcHist([hsv_blue], [0], None, [256], [0, 256])
plt.subplot(2,4,2)
plt.plot(hist1)


_, image_binary = cv2.threshold(hsv_blue, 108, 255, cv2.THRESH_BINARY) ## (threshold, image)
plt.subplot(2,4,3)
plt.imshow(image_binary,cmap="gray")
plt.title("threshold")
plt.axis("off")

hist2 = cv2.calcHist([image_binary], [0], None, [256], [0, 256])
plt.subplot(2,4,4)
plt.plot(hist2)

# Try to use erosion, dilation, open, close to fill up the lines.
kernel = np.ones((5, 5), np.uint8)
dilated_img = cv2.dilate(image_binary, kernel, iterations=3)
plt.subplot(2,4,5)
plt.imshow(dilated_img,cmap="gray")
plt.title("dilated_img")
plt.axis("off")

eroded_img = cv2.erode(dilated_img, kernel, iterations=4)
plt.subplot(2,4,6)
plt.imshow(eroded_img,cmap="gray")
plt.title("eroded_img")
plt.axis("off")

kernel = np.ones((3, 3), np.uint8)
dilated_img = cv2.dilate(eroded_img, kernel, iterations=3)
plt.subplot(2,4,7)
plt.imshow(dilated_img,cmap="gray")
plt.title("dilated_img")
plt.axis("off")


# save result
cv2.imwrite('output/morpho_blue.jpg', dilated_img)
```

## **6. Find countours**

Use package to find the contours in the image.

Practice:
1. Find the pad and probemark contours in the pad series images.
2. Calculate the countours area.

```python
pad_task1 = cv2.imread('input/binary_task1.bmp')
ori_gray = cv2.cvtColor(pad_task1, cv2.COLOR_BGR2GRAY)
plt.subplot(1,2,1)
plt.imshow(ori_gray,cmap="gray")
plt.title("origin")
plt.axis("off")

# Find the pad and probemark contours and draw.

_, image_binary = cv2.threshold(pad_task1, 100, 255, cv2.THRESH_BINARY) ## (threshold, image)
img_gray = cv2.cvtColor(image_binary, cv2.COLOR_BGR2GRAY)
contours, hierarchy = cv2.findContours(img_gray, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)  #cv2.RETR_EXTERNAL #cv2.RETR_TREE

cv2.drawContours(pad_task1, contours, -1, (255,0,0),1)
plt.subplot(1,2,2)
plt.imshow(pad_task1)
plt.title("counter")
plt.axis("off")


pad_area = []
for i in range(len(contours)):
    area = cv2.contourArea(contours[i])
    pad_area.append(area)

# save and print result
cv2.imwrite('output/pad_task1_cont.bmp', pad_task1)
print('The pad area is ', pad_area)
# print('The probe mark area is ', probemark_area)
```
